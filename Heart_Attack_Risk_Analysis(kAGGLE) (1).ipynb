```python
df_train = df_train.drop(['Patient ID', 'Blood Pressure'], axis=1)

columns_to_standardize = ['Age', 'Cholesterol', 'Heart Rate', 'Exercise Hours Per Week',
                              'Stress Level', 'Sedentary Hours Per Day', 'Income', 'BMI',
                              'Triglycerides', 'Physical Activity Days Per Week', 'Sleep Hours Per Day']
scaler = StandardScaler()
df_train[columns_to_standardize] = scaler.fit_transform(df_train[columns_to_standardize])

columns_to_encode = ['Sex', 'Diet', 'Country', 'Continent', 'Hemisphere']
df_train = pd.get_dummies(df_train, columns=columns_to_encode, drop_first=True)
```

```python
train_df, val_df = train_test_split(df_train, test_size=0.2)
```

```python
X_train = train_df.drop(['Heart Attack Risk'], axis=1)
y_train = train_df['Heart Attack Risk']

X_val = val_df.drop(['Heart Attack Risk'], axis=1)
y_val = val_df['Heart Attack Risk']
```

```python
#import
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
```

```python
random = RandomForestClassifier(ccp_alpha=0.001, random_state=1, criterion='gini')
random.fit(X_train, y_train)
y_pred = random.predict(X_val)

y_pred
```

```python
accuracy = accuracy_score(y_val, y_pred)
accuracy
```
